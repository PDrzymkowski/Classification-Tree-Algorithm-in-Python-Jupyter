### a) Wczytanie danych
Pierwszym krokiem realizacji projektu jest wczytanie danych z pliku *agaricus-lepiota.data* i zapisaniu przykładów podanych w postaci wartości atrybutów przedzielonych przecinkami do konkretnej struktury danych. W projekcie zdecydowano się na zastosowanie **listy słowników**, której każdy element zawiera (jako klucz) nazwę atrybutu i wartość ze wspomnianego pliku. Efektem tego działania jest prosty zbiór umożliwiający łatwe odwoływanie się do każdego przykładu oraz jego atrybutów. Tak stworzona struktura widnieje w projekcie pod nazwą X. Uby uzyskać lepszą różnorodność danych struktura X została przemieszana funkcją **shuffle** z modułu **random**. Przyczyną takiej decyzji był fakt, że niektóre następujące po sobie przykłady w zbiorze mają tę samą wartość konkretnego atrybutu (nawet kilkaset przykładów pod rząd), co moze wprowadzać pewne błedy w przypadku wyboru zbiorów uczących i trenujących.

Drugim krokiem było stworzenie listy atrybutów **list_of_attributes** służącej w późniejszej części projektu jako zbiór możliwych testów **S**.

```python
from random import shuffle
file = open('agaricus-lepiota.data')
X = []
for line in file:
    strings = line.split(',')
    strings[len(strings)-1] = strings[len(strings)-1].strip('\n')
    x = {'edibility':   strings[0],
         'cap-shape':   strings[1],
         'cap-surface': strings[2],
         'cap-color':   strings[3],
         'bruises':    strings[4],
         'odor':       strings[5],
         'gill-attachment':   strings[6],
         'gill-spacing': strings[7],
         'gill-size': strings[8],
         'gill-color': strings[9],
         'stalk-shape': strings[10],
         'stalk-root': strings[11],
         'stalk-surface-above-ring': strings[12],
         'stalk-surface-below-ring': strings[13],
         'stalk-color-above-ring': strings[14],
         'stalk-color-below-ring': strings[15],
         'veil-type': strings[16],
         'veil-color': strings[17],
         'ring-number': strings[18],
         'ring-type': strings[19],
         'spore-print-color': strings[20],
         'population': strings[21],
         'habitat': strings[22]}
    X.append(x)

shuffle(X)    

list_of_attributes = [ 'cap-shape',
'cap-surface',
'cap-color',
'bruises',
'odor',
'gill-attachment',
'gill-spacing',
'gill-size',
'gill-color',
 'stalk-shape',
'stalk-root',
'stalk-surface-above-ring',
'stalk-surface-below-ring',
'stalk-color-above-ring',
'stalk-color-below-ring',
'veil-type',
'veil-color',
'ring-number',
'ring-type',
'spore-print-color',
'population',
'habitat']
```

### b) Funkcje operujące na zbiorze danych
W projekcie wymagane były funkcje umożliwiające charakteryzację, podział i klasyfikację podanego zbioru danych. Przy ich pomocy możliwe będzie m.in. wyznaczenie zysku informacyjnego zbioru trenującego, możliwość podziału tegoż zbioru ze względu na wartości konkretnego parametru, określenie unikalnych wartości w zbiorze, co przydatne okaże się podczas tworzenia gałęzi z korzenia, czy wreszcie możliwość predykcji etykietowania na danym zbiorze.

W celu realizacji tych założeń zaimplementowane zostały funkcje:

+ __unique_vals(data_set, atribute)__ - Znajduje unikalne wartości atrybutów w zbiorze trenującym poprzez zapisanie do zbioru (struktura 'set') unikalnych wartości atrybutu **attribute** dla zadanych przykładów z **data_set**
+ __class_val_count(data_set)__ - Oblicza liczbę wystąpień instancji danej klasy wśród przykładów. Funkcja zwraca słownik o dwóch kluczach  __p__ , którego wartością jest liczba przykładów dla klasy **trujący** i **e**  dla jadalnych przykładów. Wartość dla konkretnego klucza zwiększa się w pętli, która sprawdza każdy przykład osobno.
+ __predict_class_label(data_set)__ - Przewiduje etykietowanie danego zbioru na podstawie przykładów w nim zawartych. W ciele funkcji realizowana jest funkcja **class_val_count** celem przypisania liczby wystąpień instancji danej klasy do konretnych zmiennych. Następnie wybór etykietowania jest realizowany **metodą ruletkową**, proporcjonalną do liczby instancji danej klasy. Polega to na wylosowaniu liczby całkowitej z zakresu liczby wszystkich przykładów z **data_set**.
+ __partition(data_set, attribute)__ - Dzieli zbiór trenujący __data_set__ na podstawie testu atrybutu __attribute__. Wynikiem działania funkcji jest słownik, w którym jako klucze ustalone są unikalne wartości danego atrybutu w zbiorze data_set, natomiast jako wartości tych kluczy widnieją listy słowników, będące listami przykładów spełniających daną wartość atrybutu



Warto zaznaczyć, że wybór etykietowania w funkcji __predict_class_label(data_set)__ nie jest ustalony *na sztywno*, lecz ustawiany z prawdopodobieństwem proporcjonalnym do liczby instancji danej klasy w zbiorze. Wynikiem takiego działania jest większa różnorodność wyboru etykietowania. Może się w końcu zdarzyć, iż podział etykietowań w danej klasie będzie w proporcji 49%:51%. W takim przypadku wybór liczniejszego zbioru jest niepewny i ryzykowny, dlatego warto go poddać rachunkowi prawdopodobieństwa.


### c) Opis elementów drzewa
Drzewo klasyfikacyjne budowane w projekcie zakłada istnienie dwóch instancji w swej strukturze: węzłów oraz liści. W węzłach realizowane są zapytania/testy dot. następnego podziału zbioru na mniejsze. W liściach zaś podejmowana jest decyzja odnośnie etykietowania. W zastosowanym rozwiązaniu obiekty węzła i liścia są instancjami klas:
+ **Node** - posiada 4 atrybuty: 
    + __T__ będący zbiorem przykładów w danym węźle (dla korzenia drzewa jest pełnym zbiorem trenujacym)
    + __test__ będący testem atrybutu wybranym w danej iteracji
    + __children__ będący listą węzłów wychodzących z gałęzi danego węzła
    + __prev_test_val__ będący wynikiem testu poprzedniego węzła (dla korzenia drzewa jest równy '')

    oraz metody:
    + __predict(self, example)__ służy sprawdzeniu, czy dany przykład ze zbioru testujacego spełnia wynik testu dla  którejś z gałęzi. W przypadku dojścia do liścia zwracane jest etykietowanie, w przypadku zatrzymania się w węźle,
       np. z powodu braku określonej wartości (gałęzi) dla danego przykładu testującego, zwracana jest 
       wartość etykietowania w postaci wyniku funkcji __predict_class_label__ której argumentem jest zbiór przykładów w danym węźle
    + __find_node_pruning(self, seeked_node)__ zamienia wskazany w argumencie węzeł na liść, co jest wymagane w funkcji przycinania drzewa. Wyszukiwanie jest wykonywane rekurencyjnie
    + __get_nodes_list(self)__ zwraca listę węzłów danego drzewa, co będzie przydatne m.in. w funkcji przycinania drzewa. Lista jest wyszukiwana rekurencyjnie, szukając dzieci w każdym węźle i wyrzucając *powtórki*

    Podczas inicjalizacji nowego węzła w jego konstruktorze podajemy parametry __T, prev_test_val__. Test oraz lista dzieci jest ustalana w trakcie działania algorytmu. Celem odwołań do poszczególnych atrybutów instacji klasy wykorzystano dekorator __properties__ języka Python.


+ **Leaf** - dziedziczy po klasie __Node__. Różnicą w jej budowie jest istnienie dodatkowego atrybutu **label** przechowującego etykietowanie danego liścia. Sposób przypisania etykiety jest realizowany za pomocą funkcji __predict_call_label__ opisanej w poprzednim podpunkcie. W przypadku reszty atrybutów należy nadmienić, że wartość __test__ dla instancji danej node jest pusta (wynosi ""), a lista __children__ także. W przypadku wywołania __get_nodes_list__ obiekt typu Leaf zwraca pustą listę, ignoruje wywołanie __find_node_pruning__, a dla wywołania __predict__ zwraca etykietowanie, czyli atrybut label. 

W dalszej części projektu klasy te i ich metody będą często wykorzystywane.

### d) Obliczenie jakości testu
Problem obliczenia jakości testu jest dość złożony i może być implementowany na wiele sposobów. W projekcie podjęto decyzję na wyznaczeniu tzw. zysku informacyjnego będącego różnicą entropii zbioru uczącego i entropii tegoż zbioru ze względu na konretny test. Entropia jest natomiast miarą nieuporządkowania zbioru, uwzględniającą m.in. jego etykietowania klasowe. W projekcie wyznaczenie zysku informacyjnego będzie miało główny udział w wyborze testu metodą ruletkową, która zostanie opisana w następnym podpunkcie.

Aby spełnić wymienione cele powstały trzy funkcje:
+ **entropy(T)** - na podstawie opisanego w dokumentacji wstępnej wzoru obliczana jest wartość entropii zbioru trenującego. Wykorzystuje do tego m.in. wywołanie funkcji __class_val_count__ 
+ **entropy_t(T,t)** - wyznacza wartość entropii zbioru trenujacego ze względu na test t. Wykorzystuje do tego m.in. wywołanie funkcji __entropy__ oraz __partition__
+ **info_gain(T, t)** -  Oblicza i zwraca zysk informacyjny zbioru uczącego T ze względu na test t wyrażony jako różnica wartości zwracanych entropy(T) od entropy_t(T,t) 

### e) Metoda ruletkowa
Znaczącym aspektem projektu jest wybór testu metodą ruletkową, tj. z prawdopodobiestwem proporcjonalnym do jakości testu. 

W funkcji **roulette(T, S)** wykorzystywane są wyniki zysku informacyjnego dla każdego testu ze zbioru możliwych testów S na zbiorze trenującym T. Aby uzyskać wynik jakości każdego testu wykonywane w pętli są obliczenia funkcji __info_gain__ 

W algorytmie każda nowa wartość zysku informacyjnego jest zapisywana do słownika **info_gain_results** pod kluczem będącym nazwą badanego testu ze zbioru S i zwiększana o wartosć z poprzedniego wyniku. Z takiego zakresu wybierana jest losowa wartość, wykorzystując funkcję *uniform* z modułu *random*. Wynikiem algorytmu jest nazwa wylosowanego testu.

### f) Algorytm budowy drzewa klasyfikacyjnego ID3
Do rozwiązania problemu budowy drzewa klasyfikacyjnego zastosowano popularny algorytm __ID3__, modyfikując go o możliwość zmiany konkretnych parametrów jego działania oraz dodając losowanie ruletkowe. Powstała dzięki temu funkcja __ID3_build_tree(T, S, test_result, Z_min, L_min, mode)__ 
Parametrami funkcji są:
+ T - zbiór uczący z dziedziny przykładów X
+ S - zbiór możliwych testów atrybutów
+ test_result - wynik poprzedniego testu celem rozróżnienia gałęzi drzewa i kosntruowania nowych węzłów/liści
+ Z_min, L_min - parametry budowy drzewa opisane poniżej
+ mode - określający, czy test ma być wybierany metodą ruletkową (wartość __True__), czy najlepszego testu (__False__)

Funkcja ma charakter rekurencyjny. W przypadku decyzji o tworzeniu węzła wykonywana jest funkcja __ID3_build_tree(T_value, S_part, value, Z_min, L_min, mode)__ zawierająca jako argumenty konkretny podzbiór T_value spełniający wynik testu, wartość testu value oraz pomniejszony o dany test zbiór S_part. 

Algorytm dokonuje kolejno sprawdzenia zawartości etykietowań każdej klasy w zbiorze T, aby ustalić, czy należy w miejsce obecnie tworzonego węzła wstawić lisć zgodny z najpospolitszym etykietowaniem. Sprawdzana też jest zawartość możliwych testów w zbiorze S. W przypadku braku decyzji o tworzeniu liścia następuje *losowanie* parametru dzielącego zbiór T i wyznaczanie kolejnych węzłów rekurencyjnie lub tworzenie liści w przypadku, gdy liczność podzbioru T__value jest mniejsza od określonej granicy.

Do sprawdzenia liczby instancji danej klasy wykorzystuje się funkcję __class_val_count__ (punkt 1b sprawozdania).  Wybór atrybutu A jest realizowany przez funkcję __roulette__. **Zauważyć można zatem, że za każdym wywołaniem funkcji *od zera* wynik może być inny ze względu na losowść wyboru testu.** Unikalne wartości dla wybranego testu są znajdowane funkcją __unique_vals__, podział na podzbiory zbioru trenujacego realizuje funkcja __partition__. Wartością zwracaną przez algorytm jest węzeł/liść drzewa decyzyjnego. Jako wynik algorytmu otrzymujemy drzewo klasyfykacyjne (a dokładniej jego korzeń) opisujące problem zadania

### g) Funkcje testujące drzewo
Aby poprawnie i efektywnie przetestować zbudowane drzewo należy dostarczyć w odpowiedni sposób dane testujące oraz przedstawić wyniki klasyfikacji. Służy do tego funkcja **test_tree(tree, test_set, printing)**, przyjmująca jako argumenty korzeń drzewa klasyfikacyjnego __tree__ zbudowanego algorytmem ID3, zbiór przykładów testujących __test_set__ z dziedziny X, a także argument typu Boolean __printing__ którego ustawienie na True wyświetla wyniki klasyfikacji na ekranie. Należy zadbać, aby zbiór __tree__ ten był rozłączny względem zbioru trenującego, nie chcemy testować drzewa na *znanych* przykładach. Algorytm ten wykorzystuje metodę obiektów klasy **Node** - **predict()**. Na podstawie wyników tej metody sprawdzana jest liczba etykietowań jadalnych grzybów oraz trujących i porównana z rzeczywistą miarą. Na końcu wyznaczany jest stosunek określony jako jakość klasyfikacji i zwracany w postaci zmiennej __error__.

Kolejną istotną funkcją programu jest implementacja tzw. *walidacji krzyżowej*. Funkcja **cross_validation(X, S, iterations, printing, Z_min, L_min, mode)** stanowi swego rodzaju środowisko testowe do budowania i testowania drzew odpowiednią, opisaną przez argument **iterations**, ilość razy. 

Opis argumentów tej funkcji jest następujacy:
+ X - zbiór wszystkich przykładów dziedziny X
+ S - zbiór możliwych testów atrybutów
+ iterations - liczba iteracji, czyli *przesunięć* zbioru testującego. Aby nie doszło do nachodzenia na siebie zbiorów w kolejnych iteracjach, a przykłady były unikalne za każdym razem wartość ta powinna się znajdować w zakresie od 1 do 7
+ printing - decyzja, czy wyświetlać informacje o jakości klasyfikacji, zmianach parametrów, czy liczbie węzłów w drzewie
+ Z_min, L_min - parametry budowy drzewa/algorytmu ID3
+ mode - decyzja, czy wybór testu budowanego drzewa ma być realizowany metodą ruletkową, czy najlepszego testu

Zbiór dziedziny **X** jest dzielony na 3 podzbiory. Jednym podzbiorem jest w tym przypadku zbiór trenujacy o wielkości ponad 6000 przykładów, znacznie większy od zbioru testującego, który stanowi drugi podzbiór (1000 przykładów).
Trzecim podzbiorem jest tzw. __zbiór przycinający__ niezależny od dwóch poprzednich i wykorzystywany przy funkcji przycinania drzewa. Funkcja kolejno buduje drzewo dla zbioru trenującego, dokonuje jego przycięcia i liczy jakość klasyfikacji na podstawie przykładów ze zbioru testującego, aby następnie *przesunąć* zakres zbiorów o odpowiednią ilość i przeprowadzić operację ponownie konkretną, opisaną przez argument funkcji __iterations__ ilość razy. 

Wynikiem działania funkcji są dwie struktury:
+ error_array - będąca listą jakości na zbiorze testujacym klasyfikacji dla każdej iteracji
+ nodes_array - będąca listą zawierającą liczby węzłów kosntruowanych dla każdej iteracji.

### h) Przycinanie drzewa

Celem uniknięcia nadmiernego dopasowania algorytmu do zbioru trenującego stosuje się metody przycinania drzew poelgające na uproszczeniu struktury drzewa poprzez zamianę części jego węzłów na liscie, bazując na wynikach jakości klasyfikacji dla niezależnego zbioru testujacego (zbioru przycinajacego). Jedno z rozwiązań tego typu zastosowano w projekcie pod funkcją o nazwie __tree_pruning(tree, data_set)__. Jej argumentami są:
+ tree - zbudowane drzewo, czyli jego korzeń główny
+ data_set - będący niezależnym zbiorem przykładów do przycinania

Sposób działania funkcji wygląda następująco:
+ wyznaczana jest jakość klasyfikacji dla obecnego drzewa __quality__
+ tworzona jest kopia oryginalnego drzewa __tree_temp__ oraz dla dupliaktu pobierana jest lista węzłów metodą __get_nodes_list__ z klasy __Node__ (p. 1c sprawozdania)
+ dla każdego węzła następuje zamiana jego klasy na lisc __Leaf__ metodą klasy Node __find_node_pruning__
+ obliczana jest na tej podstawie jakość klasyfikacji dla tak zmodyfikowanego drzewa __quality_temp__
+ jeśli __quality_temp > quality__ to następuje skopiowanie nowo powstałego drzewa przez oryginalne
+ jeśli nie, to drzewo __tree_temp__ kopiuje oryginalne 
+ argumentem zwracanym przez funkcję jest korzeń drzewa __tree__

Funkcja wykorzystuje moduł **copy** do tworzenia głebokich kopii obiektów typu __Node__ funkcją __deepcopy__

### a) Zmiana parametru Z_min przy *dobrej* wartości L_min

Pierwszym eksperymentem bedzie sprawdzenie jak na jakość oraz rozmiar drzewa wpłynie zmiana parametru Z_min decydującego na początku działania algorytmu o __stworzeniu liścia, jeśli odsetek instancji którejś z klas będzie większy niż Z_min__. Przwidujemy, iż jakość oraz rozmiar drastycznie spadną dla niskich wartości tegoż parametru. Jego zmian będziemy dokonywać w zakresie __od 0.49 do 1.00 z krokiem 0.1 dla 7 iteracji walidacji krzyżowej__. Parametrem, który nie będzie zmieniany w tym przypadku jest **L_min**, którego wartość ustalamy na *dobrą*, czyli stosunkowo niewielką, bo wynoszącą **50** przykładów w węźle. Opis wpływu zmian L_min zostanie przedstawiony w następnym podpunkcie.

Jako parametr __mode__ funkcji __ID3_build_tree__ w następnych 6 podpunktach przyjęto wartość __True__, aby korzystać z metody ruletki przy wyborze testu (parametr ustawiany w funkcji __cross_validation__. Parametr __printing__ został ustalony na __False__, aby nie wyświetlać wyników klasyfikacji dla każdej iteracji (gdyż zajęłoby to sporo miejsca).

```python
from numpy import arange
import Projekt

# Inicjalizacja badanych parametrów oraz struktur do przechowywania danych
Z_min = 0.99
L_min = 50
errors_Z_min = []
tree_nodes_Z_min = []
Z_min_change = []

for Z_min in arange(0.49, 1.00, 0.1):
    #print('\nEksperyment dla Z_min = %.2f' % Z_min)
    
    #Przerowadzenie walidacji krzyżowej, która jednocześnie tworzy, przycina i testuje drzewo
    #Wartość printing = False, natomiast mode = True
    errors_temp, nodes_temp = Projekt.cross_validation(X, S, 6, False, Z_min, L_min, True)
    #Zapisanie wyników do odpowiednich struktur danych
    errors_Z_min.append(errors_temp)
    tree_nodes_Z_min.append(nodes_temp)
    Z_min_change.append(Z_min)
    
```

#### Wnioski:
+ Zarówno jakość klasyfikacji jak i rozmiar drzewa są silnie zależne od parametru __Z_min__, jego zbyt mała wartosć ma przełożenie na małą wielkość drzewa, a co za tym idzie znikomą jakość klasyfikacji
+ Zbyt małe drzewo nie posiada wielu gałęzi, po których możnaby *dopasować* przykłady zbioru testującego. Decyzja o budowie liścia jest w tym przypadku podejmowana __na początku działania algorytmu__, gdy sprawdzany jest odsetek przykładów instancji każdej klasy. W przypadku gdy __Z_min__ jest bliski 50% budowany jest tylko jeden liść będący jednocześnie korzeniem całego drzewa. Jeśli zakładamy, że zbiór testujacy statystycznie składa się w połowie z instancji jednej klasy i w połowie z drugiej to jakość wyniesie ok 50% - __co też zostało udowodnione na wykresie__
+ Wraz ze wzrostem parametru Z_min __rośnie jakość klasyfikacji__, aż osiąga kulminację w okolicach 100% w przypadku, gdy Z_min = 0.99. W takim przypadku, aby stworzyć nowy liść algorytm musi być *prawie* pewien, że w danej klasie są prawie same przykłady danej klasy. Skutkiem ubocznym zysku na jakości jest natomiast __znaczny wzrost złożoności drzewa__. Liczba węzłów wzrosła diametralnie, nawet mimo oepracji przyciania drzewa, co negatywnie wpłyneło na czas wykonywania się programu i jego złożonosć obliczeniową i pamięciową.
+ Warto zaznaczyć, że wybór etykietowania liścia ma w pewnym sensie charakter losowy (proporcjonalny do ilosci przykładów każdej klasy), więc ewentualne spadki jakości klasyfikacji mogą wynikać z takiego zabiegu.
+ Wpływ rouletkowego wyboru testu też nie jest bez znaczenia. Wprowadzona losowość może miejscami zaburzać wyniki, dodając element niepewności, co do wyboru testu dla konkretnego podziału. Dokładny wpływ tego zjawiska zostanie opisany w ostatnim podpunkcie.

```python

# Inicjalizacja badanych parametrów oraz struktur do przechowywania danych
L_min_change = [800, 650, 500, 350, 200, 50]
Z_min_change = [0.49 ,0.59, 0.69, 0.79, 0.89, 0.99]

errors=[]
tree_nodes= []


for n in range(6):
    
    L_min = L_min_change[n]
    Z_min = Z_min_change[n]
    
    #print('\nEksperyment dla L_min = %d i Z_min = %d procent' % (L_min, Z_min*100))
    
    
    #Przerowadzenie walidacji krzyżowej, która jednocześnie tworzy, przycina i testuje drzewo
    #Wartość printing = False, natomiast mode = True
    errors_temp, nodes_temp = Projekt.cross_validation(X, S, 6, False, Z_min, L_min, True)
    #Zapisanie wyników do odpowiednich struktur danych
    errors.append(errors_temp)
    tree_nodes.append(nodes_temp)
```

```python
#Przemnożenie danych, aby możliwe było ich wyświetlenie w procentach
Z_min_change_percent = [int(z_min*100) for z_min in Z_min_change]

#Wyświetlanie wykresów jakości klasyfikacji plot i boxplot
plt.plot(Z_min_change_percent, errors, 'ro')
plt.grid(True)
plt.xlabel('Z_min [%] | L_min = [750, 500, 350, 200, 50]')
plt.ylabel('Jakość [%]')
plt.title('Zależność jakości klasyfikacji od Z_min i L_min')
plt.show()

plt.boxplot(errors, positions = Z_min_change_percent )
plt.grid(True)
plt.xlabel('Z_min [%] | L_min = [750, 500, 350, 200, 50]')
plt.ylabel('Jakość [%]')
plt.title('Zależność jakości klasyfikacji od Z_min i L_min')
```

```python
#Wyświetlanie wykresów ilości węzłów plot i boxplot
plt.plot(Z_min_change_percent, tree_nodes, 'ro')
plt.grid(True)
plt.xlabel('Z_min [%] | L_min = [750, 500, 350, 200, 50]')
plt.ylabel('Liczba węzłów')
plt.title('Zależność liczby węzłów drzewa klasyfikacji od Z_min i L_min')
plt.show()

plt.boxplot(tree_nodes, positions = Z_min_change_percent )
plt.grid(True)
plt.xlabel('Z_min [%] | L_min = [750, 500, 350, 200, 50]')
plt.ylabel('Liczba węzłów')
plt.title('Zależność liczby węzłów drzewa klasyfikacji od Z_min i L_min')
```

#### Wnioski:
+ Obserwujemy niską jakość klasyfikacji oraz niewielkie rozmiary drzew dla przypadku *kiepskich* wartości parametrów L_min i Z_min. Załozenia i przewidywania zostały potwierdzone: __wraz ze wzrostem Z_min i spadkiem L_min obserwujemy poprawę działania algorytmu oraz zwiększenie złożoności budowanych drzew.__ 
+ W przypadku małych wartości, gdy __Z_min wynosi ok. 0.5__ wartość L_min nie ma większego znaczenia, gdyż algorytm tworzy tylko jeden węzeł (Liść). Jakość jest wtedy równa ok. 50%
+ W przypadku, gdy oba parametry osiagaję kulminację (L_min = 50, Z_min = 0.99) obserwujemy niemalże idealną klasyfikację dla wszystkich danych testowych, czego skutkiem ubocznym jest znaczne zwiększenie wielkosci drzewa

W przypadku metody najlepszego testu, wpływ opisywanych parametrów nie jest już tak silny, jakość dalej jest utrzymywana dla wszystkich większości testów na poziomie powyżej 95%.
+ Metoda ruletkowa jest bardzo podatna na wpływ podanych parametrów. Nie mamy pewności, czy wybrany test rzeczywiście dokonuje najlepszego możliwego podziału zbioru trenującego, lecz jest to realizowane w sposób losowy. Mimo zależnosci wyboru od jakości testu istnieje szansa, że wybrany zostanie test bardzo nieznacznie dzielący dany zbiór na podzbiory.
+ Warto zwrócić jednak uwagę, że dla 2giej iteracji w przypadku 3ego testu (dla słabych parametrów) metoda ruletkowa poradziła sobie o wiele lepiej od metody najlepszego testu. Losowanie w tym przypadku dało pozytywny skutek na jakość klasyfikacji. 
+ Jednocześnie nie sposób nie zauważyć sytuacji generowania jedynie jednego węzła dla gorszych wartości parametrów. Dla metody najlepszegot testu już w drugim teście (dla *średnich* parametrów) uzyskujemy po jednym węźle w każdej iteracji. Świadczyć to może o dobrym podziale zbioru trenującego na odpowiednie podzbiory, których wielkosć lub odsetek etykietowań danej klasy mieści się w ograniczeniach nałożony przez parametry (stąd  budowa liści). 
+ Inną przyczyną takiej sytuacji może być także __operacja przycinania drzewa__, która w obu metodach ogranicza wielkość drzewa. Przycinanie drzewa w teorii ma za zadanie ograniczyć ilość węzłów poprzez zastąpienie ich liścmi tak, aby dla zadanego zbioru jakosć klasyfikacji była lepsza. Należy jednak nadmienić, że etykietowanie tak stworzonego liscia jest realizowane w sposób proporcjonalny do ilości instancji danej klasy w liściu. Dodaje nam to charakteru losowości w procesie klasyfikacji. Operacja przycinania nie podlega już bowiem ograniczeniom związanym z parametrami __L_min i Z_min__, więc jest dość spora szansa na stworzenie liścia o liczbie przykładów równej nawet kilu tysiącam. Dodając do tego charakter losowości spowodowany ruletkowym wyborem testu wyniki mogą odbiegać od idealnych.
+ Skutkiem tego typu zachowań i charakterystyk metody ruletkowej jest znaczne zwiększenie liczby węzłów oraz osłabienie jakości klasyfikacji względem metody najlepszego testu

## 3. Podsumowanie
Udało się zrealizować algorytm budowania drzewa klasyfikacyjnego metodą ruletkową. Zaimplementowano potrzebne funkcje i stworzono środowisko do testów zmian parametrów badanego algorytmu. Na koniec porównano metodę ruletkową z metodą najlepszego testu. Na podstawie wszystkich wyników można wysunąć następujace wnioski:
+ Drzewo klasyfikacyjne pozwala z sukcesem zrealizować rozwiązanie problemu efektywnej klasyfikacji przy jednoczesnej przejryzstej prezencji uzyskanych wyników. Wybór metody wyboru testu oraz parametrów kryterium stopu w algorytmie budowy drzew decyzyjnych stanowi natomiast jedno z ważniejszych zagadnień dot. uzyskania dużej jakości klasyfikacji.
+ Metoda ruletkowa wprowadza dużą losowość do generowanych wyników. Niemniej, jesteśmy w stanie zauważyć zależność jakości klasyfikacji oraz wielkości budowy drzewa od zadanych parametrów algorytmu.
+ W przypadku dobrych wartości parametrów (Z_min bliskie 1, L_min odpowiednio małe) uzyskujemy bardzo dobrą jakość klasyfikacji, kosztem wielkości drzewa i jego złożoności 
+ Parametr Z_min opisujący min. odsetek instancji danej klasy w zbiorze aby utworzyć liść ma decydujące znaczenie w przypadku jakości i złożoności budowanego drzewa. 
+ Parametr L_min ma natomiast duży wpływ na ilosć budowanych węzłów i niewielki wpływ na jakość klasyfikacji. Połączenie dużej wartości L_min z dużą wartością Z_min pozwolić może na otrzymanie dobrej jakości klasyfikacji przy niewielkim rozmiarze drzewa.
+ Metoda najlepszego testu okazała się skuteczniejsza pod względem jakości i niewrażliwa na mocne zmiany parametrów
+ Metoda ruletkowa jest przydatna w przypadku, gdy chcemy wygenerować odpowiednią losowość budowania drzewa. Może się to okazać w przypadku, gdy chcemy przebadać algorytm pod względem różnych parametrów lub, gdy zależy nam na różnorodnośi budowania drzew decyzyjnych. Metoda najlepszego testu bowiem nie zawsze musi zwrócić test pasujący do klasyfikacji zbioru testujacego (co widać dla 3ego testu w podpunkcie 2f, gdzie metoda ruletkowa okazała się lepsza w jednej iteracji). Dzięki metodzie ruletkowej uzyskujemy szansę na wybór mniej oczywistego testu pod względem jakości, ale być moze lepiej zareaguje on na różnorodne dane testowe.
+ Metoda ruletkowa wyboru testu może zatem posłużyć do większego zabezpieczenia się przed nadmiernym dopasowaniem do zbioru uczącego, czego nie można powiedzieć o metodzie najlepszego testu. Należy jednak uważać z jej implementacją, gdyż może (m.in. w połączeniu z algorytmem przycinania drewa) wprowadzać duże niepewności dot. klasyfikacji.
